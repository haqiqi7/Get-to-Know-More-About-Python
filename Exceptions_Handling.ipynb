The exception handling process uses a try statement paired with except

**ZeroDivisionError** : It Occurs when a number is divided by zero.

```python
x = 0
```

```python
division = 50/x
```

Handling Excetions for ZeroDivisionError is done so that the application no longer exists execution due to an error, but is replaced by printing a message to the screen.


```python
#solve

x = 0

try:
  div = 50/x
  print(div)
except ZeroDivisionError:
  print("can't divide a number by zero")
```

**FileNotFound** : It occurs when a file is not found. It may be local or global.

```python
open('empty.py')
```

```python
#solve 

try:
  with open('empty.py') as file:
    print(file.read())
except FileNotFoundError:
  print("Not Found 'empty.py'")
```

**FileNotFoundError** Exception Handling as a one-element tuple, don't forget to write a one-element tuple must still end with a comma.

```python
#solve 

try:
  with open('empty.py') as file:
    print(file.read())
except (FileNotFoundError, ):
    print("Not Found empty.py")
```

**KeyError** it occurs when you try to access a key that isnâ€™t in a dictionary (dict).

```python
d = {'sum':'5.0'}

print('sum: {}'.format(d['Sum']+3))
```

in more complex applications, exception handling can use a single except statement which handles more than one error type combined in a tuple

```python
d = {'sum' : '5.0'}

try:
  print('sum: {}'.format(d['Sum']+3))
except KeyError:
  print('key not found in dictionary')
except (ValueError, TypeError):
  print('Invalid value or type')
```

```python
x = {'sum': '5.0'}

try:
  print('total result: {}'.format(int(x['sum'])))
except (ValueError, TypeError) as y:
  print('handling eror: {}'.format(y))
```
